```python
data["ingredient_name"]=data["ingredient_name"].apply(
    lambda lst: list(dict.fromkeys(
        item.strip().lower() for item in lst
    ))
)
```

```python
data.loc[data["ingredient_name"].apply(len).idxmax()]
```

```python
data["ingredient_name"] = data["ingredient_name"].apply(
    lambda x: sorted(x, key=str.lower)
)
```

```python
ingredients = [
    "Basmati Rice", "Cashews", "Chickpea Flour", "Cloves", "Cream",
    "Fenugreek Leaves", "Fish", "Garlic", "Ghee", "Ginger",
    "Lemon Juice", "Mustard Seeds", "Onion", "Potato", "Raisins",
    "Salt", "Sugar", "Tofu", "Vinegar"
]
```

```python
df=data.copy()
```

```python
df.head()
```

```python
unique_ingredients = sorted(
    df["ingredient_name"].explode().unique()
)

print(unique_ingredients)
print(len(unique_ingredients))
```

```python
for ingredient in unique_ingredients:
    df[ingredient] = df["ingredient_name"].apply(
        lambda x: int(ingredient in x)
    )
```

```python
df=df.drop(["recipe_id","ingredient_name"],axis=1)
```

```python
df.head()
```

```python
df.to_csv("Recipe_ingrediect.csv", index=False)
```
